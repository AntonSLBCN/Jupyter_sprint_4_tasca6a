Treballa plot donuts amb diferentes parts de dataset en aquest cas cada 50 anys

```python
movies_definitiu['Years'] = pd.to_numeric(movies_definitiu['Years'])
```

```python
df19001950 = movies_definitiu.loc[movies_definitiu['Years'] <= 1950]
df19001950 = df19001950.iloc[ : , 3:22]
df19001950 = df19001950.groupby(['Years']).sum()
df19512000 = movies_definitiu.loc[movies_definitiu['Years'] >= 1951]
df19512000 = df19512000.iloc[ : , 3:22]
df19512000 = df19512000.groupby(['Years']).sum()

#convertir en lista
L19001950 = [df19001950[x].sum() for x in df19001950.columns]
L19512000 = [df19512000[x].sum () for x in df19512000.columns]
```

```python
fig1_1 = plt.figure(figsize=(10,10))
fig1_1.tight_layout()
plt.title('Generos entre 1900 y 1950')
plt.pie(L19001950,autopct='%1.1f%%', shadow=False, startangle=90, labeldistance=0.5 )
plt.style.use('ggplot')
plt.legend(list(df19001950), loc='center left',  bbox_to_anchor=(1, 0, 0.5, 1))
sfera = plt.Circle(xy=(0,0),radius= 0.40, facecolor='white')
plt.gca().add_artist(sfera)
plt.show
```

```python
fig1_2 = plt.figure(figsize=(10,10))
fig1_2.tight_layout()
plt.style.use('ggplot')
plt.title('Generos entre 1951 y 2000')
plt.pie(L19512000,autopct='%1.1f%%', shadow=False, startangle=90, labeldistance= 10 )
plt.legend(list(df19512000), loc='center left',  bbox_to_anchor=(1, 0, 0.5, 1))
sfera = plt.Circle(xy=(0,0),radius= 0.40, facecolor='white')
plt.gca().add_artist(sfera)
plt.show
```
